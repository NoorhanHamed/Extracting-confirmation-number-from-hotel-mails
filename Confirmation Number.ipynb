```python
df = df[['subject', 'content', 'ticket_id', 'category','status_y']]
df['content'] = df['content'].apply(lambda x:x.replace(' , ','-'))
df.head()
```

```python
text_of_refs = ['hcn','conf','confirmation','confirmations','ref', 'reconfirmation']
stop_W = ['.',',', ';', "\'",":","no",'\t','(',')','_', 'booking','is', 'number','numbers'
          ,'for','the','below',"#","â€œ","code" ,"for", "room",']','[','dbl.gv','tpl.om']
w_spaces = [' ', '  ', '   ', '\t','           ','        ','\n']
replaces = ["x000d"]
```

```python
def list_of_tokens(String):
    doc = nlp(String)
    
    for token in doc:
        list_tokens.append(token.text)
        
    return list_tokens    
```

```python
def remove_stop_words(list_items, stop_W):
    new_list = []
    for item in list_items:
        item = item.strip()
        if (item) and (item not in stop_W):
            new_list.append(item)
    return new_list
```

```python
def confirmed_number_identifier(final_list):
    for key,token in enumerate(final_list):
        if ((token in text_of_refs) and (len(final_list[key+1]) > 4)):
                conf_nums.append(final_list[key+1])
                if token == text_of_refs[0]:
                    break
    return conf_nums    
```

```python
def cleaning_content(df):
    for reg in replaces:
        df['content'] = df['content'].str.replace(reg," ")
    return df    
```
